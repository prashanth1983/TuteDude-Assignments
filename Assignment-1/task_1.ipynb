#### List

```python
products = ["xps-laptop", "inspiron-laptop", "gaming-latop", "alienware-laptop","chromebook", "workstation-laptop"]
print(products[1])
print(products[-1])


products.append("tablet")
products.append("smartphone")
print(products)

# create tuple
sample_products = ("xps-laptop", 100,"laptop")
print(sample_products)

# convert tuple to list to modify
sample_list=list(sample_products)
print(sample_list)
sample_list[1]=120
print(sample_list)

# convert back to tuple
sample_products=tuple(sample_list)
print(sample_products)
```
