### Dictionary

```python
products = ["xps-laptop", "inspiron-laptop", "gaming-latop", "alienware-laptop","chromebook", "workstation-laptop"]
price_dict={
    "xps-laptop": 100,
    "inspiron-laptop": 200,
    "gaming-laptop": 800,
    "alienware-laptop": 2000,
    "chromebook": 150,
    "workstation-laptop": 3000}

print(price_dict)

# add a new product to the dictionary
price_dict["tablet"] = 750
price_dict.update({"smartphone": 600})

# update the price of an existing product
price_dict["gaming-laptop"] = 850

print(price_dict)

#remove a product from the dictionary
del price_dict["chromebook"]
price_dict.pop("inspiron-laptop")

print(price_dict)


# average price of all products
total_price = sum(price_dict.values())
average_price = total_price / len(price_dict)
print(f"Average price of all products: {average_price}")


max_price_product = max(price_dict, key=price_dict.get)
min_price_product = min(price_dict, key=price_dict.get)

print(f"Most expensive product: {max_price_product} at price {price_dict[max_price_product]}")
print(f"Least expensive product: {min_price_product} at price {price_dict[min_price_product]}")
```
