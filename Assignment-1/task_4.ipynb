```python
products = ["xps-laptop", "inspiron-laptop", "gaming-laptop", "alienware-laptop","chromebook", "workstation-laptop"]
price_dict={
    "xps-laptop": 100,
    "inspiron-laptop": 200,
    "gaming-laptop": 800,
    "alienware-laptop": 2000,
    "chromebook": 150,
    "workstation-laptop": 3000}
categories = ["premium", "budget", "budget", "student", "business", "workstation"]

# using the products and price_dict , create a list of tuples where each tuple contains the product name and its price and category
catalog = []   
for i, product in enumerate(products):
    print(i, product)
    catalog.append((product, price_dict[product], categories[i]))  

print(catalog)

# from catalog, create  a disctionary that maps each category list of product names in that category
category_dict = {}
for item in catalog:
    product_name = item[0]
    product_category = item[2]
    if product_category not in category_dict:
        category_dict[product_category] = []
    category_dict[product_category].append(product_name)

print(category_dict)

# prinnt all products that belong to the category that has maximum number of products
max_category = max(category_dict, key=lambda k: len(category_dict[k]))
print(f"Category with maximum products: {max_category}")
```
